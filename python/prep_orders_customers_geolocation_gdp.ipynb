```python
import pandas as pd

prep_orders=pd.read_csv('./data/brazilian_e-commerce/prep_orders.csv')

prep_orders.head()
```

```python
prep_orders.tail()
```

```python
prep_orders.info()
```

```python
prep_customers=pd.read_csv('./data/brazilian_e-commerce/prep_customers.csv', dtype={'customer_zip_code_prefix': str})

prep_customers.head()
```

```python
prep_customers.tail()
```

```python
prep_customers.info()
```

```python
prep_customers['customer_zip_code_prefix'] = (
    prep_customers['customer_zip_code_prefix']
    .astype(str)
    .str.zfill(5)
)
```

```python
prep_customers.info()
```

```python
prep_orders_customers = pd.merge(
    prep_orders,
    prep_customers,
    how="left",
    on="customer_id"
)
prep_orders_customers.head()
```

```python
prep_geolocation=pd.read_csv('./data/brazilian_e-commerce/prep_geolocation.csv', dtype={'geolocation_zip_code_prefix': str})

prep_geolocation.head()
```

```python
prep_geolocation.tail()
```

```python
prep_geolocation.info()
```

```python
# Merge customers + geolocation on ZIP prefix
prep_orders_customers_geolocation = pd.merge(
    prep_orders_customers,
    prep_geolocation,  # cleaned geolocation table
    how="left",
    left_on="customer_zip_code_prefix",
    right_on="geolocation_zip_code_prefix"
)
prep_orders_customers_geolocation.head()
```

```python
prep_orders_customers_geolocation.info()
```

```python
prep_orders_customers_geolocation = prep_orders_customers_geolocation.drop (columns = [
    'customer_city', 
    'geolocation_zip_code_prefix',
    'geolocation_city',
    'geolocation_city_clean',
    'geolocation_state'
    ])
prep_orders_customers_geolocation.head()
```

```python
prep_gdp=pd.read_csv('./data/brazilian_e-commerce/prep_gdp.csv')

prep_gdp.head()
```

```python
prep_gdp.tail()
```

```python
prep_gdp.info()
```

```python
prep_orders_customers_geolocation_gdp = pd.merge(
    prep_orders_customers_geolocation,
    prep_gdp,  # cleaned geolocation table
    how="left",
    left_on="customer_state",
    right_on="state_code"
)
prep_orders_customers_geolocation_gdp.head()
```

```python
prep_orders_customers_geolocation_gdp.columns
```

```python
prep_orders_customers_geolocation_gdp = prep_orders_customers_geolocation_gdp.drop (columns = [
    'state_code', 
    'total_customers',
    'customers_per_million',
    'gdp_per_capita',
    ])
prep_orders_customers_geolocation_gdp.head()
```

```python
# Convert all column names to lowercase
prep_orders_customers_geolocation_gdp.columns = prep_orders_customers_geolocation_gdp.columns.str.lower()

# Verify
print(prep_orders_customers_geolocation_gdp.columns)
```

```python
prep_orders_customers_geolocation_gdp.head()
```

```python
prep_orders_customers_geolocation_gdp.to_csv('prep_orders_customers_geolocation_gdp.csv', index=False)
```

```python
from sqlalchemy import create_engine, types
from sqlalchemy import text # to be able to pass string
```

```python
# Let's load values from the .env file
from dotenv import dotenv_values

config = dotenv_values()

# define variables for the login
pg_user = config['POSTGRES_USER']  # align the key label with your .env file !
pg_host = config['POSTGRES_HOST']
pg_port = config['POSTGRES_PORT']
pg_db = config['POSTGRES_DB']
pg_schema = config['POSTGRES_SCHEMA']
pg_pass = config['POSTGRES_PASS']
```

```python
# updating the url
url = f'postgresql://{pg_user}:{pg_pass}@{pg_host}:{pg_port}/{pg_db}' #the same like version 1

# let's switch the logging off again. .
engine = create_engine(url, echo=False) #the same like version 1


# writing dataframe to DB : Pandas Dataframe to DB Table in my own Schema
prep_orders_customers_geolocation_gdp.to_sql(name = 'prep_orders_customers_geolocation_gdp', 
                       con = engine, 
                       schema = pg_schema, # pandas is allowing to specify, in which schema the table shall be created
                       if_exists='replace', 
                       index=False
                      )
```
