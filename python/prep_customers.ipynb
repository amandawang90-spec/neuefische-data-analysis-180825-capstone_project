```python
import unicodedata

def has_accent(s):
    return any(unicodedata.category(c) == 'Mn' for c in unicodedata.normalize('NFD', s))

df_customers['has_accent'] = df_customers['customer_city'].apply(has_accent)

df_customers['has_accent'].value_counts()
```

```python
df_customers['customer_city'].nunique()
```

```python
df_customers['customer_city'].sort_values(ascending=True).head(50)
```

```python
df_customers['customer_city_clean'] = (
    df_customers['customer_city']
    .str.lower()
    .str.strip()
)
df_customers
```

```python
df_customers['customer_city_clean'] = df_customers['customer_city_clean'].str.replace(r'[^a-zA-Z\s]', '', regex=True)
df_customers['customer_city_clean'] = df_customers['customer_city_clean'].str.replace(r'\s+', ' ', regex=True)
df_customers
```

```python
df_customers = df_customers.drop(columns=['has_accent'])
df_customers
```

```python
df_customers = df_customers.sort_values(by='customer_zip_code_prefix', ascending=True)
df_customers
```

```python
df_states=pd.read_csv('./data/brazilian_e-commerce/states.csv')
df_states.head()
```

```python
df_customers_states = pd.merge(
    df_customers,
    df_states,
    how="left",
    left_on="customer_state",
    right_on="UF"
)
df_customers_states.head()
```

```python
df_customers_states = df_customers_states.drop (columns = [
    'UF', 
    'Area',
    'Population',
    'Demographic Density',
    'Cities count',
    'GDP',
    'GDP rate',
    'Poverty',
    'Latitude',
    'Longitude'
    ])
df_customers_states.head()
```

```python
df_customers_states.columns= df_customers_states.columns.str.lower()
df_customers_states.head()
```

```python
df_customers_states = df_customers_states.rename(columns={
    'customer_city' : 'customer_city_old',
    'customer_state': 'customer_state_code',
    'state': 'customer_state',
    'capital': 'customer_capital',
    'region': 'customer_region'
})
df_customers_states.head()
```

```python
df_customers_states.tail()
```

```python
new_order = [
     'customer_id',
     'customer_unique_id',
     'customer_zip_code_prefix',
     'customer_city_old',
     'customer_city_clean',
     'customer_state_code',
     'customer_state',
     'customer_capital',
     'customer_region'
 ]

df_customers_states = df_customers_states[new_order]

df_customers_states.head()
```

```python
df_customers_states = df_customers_states.apply(lambda x: x.str.lower() if x.dtype == "object" else x)
df_customers_states.head()
```

```python
df_customers_states.to_csv('prep_customers.csv', index=False)
```

```python
from sqlalchemy import create_engine, types
from sqlalchemy import text # to be able to pass string
```

```python
# Let's load values from the .env file
from dotenv import dotenv_values

config = dotenv_values()

# define variables for the login
pg_user = config['POSTGRES_USER']  # align the key label with your .env file !
pg_host = config['POSTGRES_HOST']
pg_port = config['POSTGRES_PORT']
pg_db = config['POSTGRES_DB']
pg_schema = config['POSTGRES_SCHEMA']
pg_pass = config['POSTGRES_PASS']
```

```python
# updating the url
url = f'postgresql://{pg_user}:{pg_pass}@{pg_host}:{pg_port}/{pg_db}' #the same like version 1

# let's switch the logging off again. .
engine = create_engine(url, echo=False) #the same like version 1


# writing dataframe to DB : Pandas Dataframe to DB Table in my own Schema
df_customers_states.to_sql(name = 'prep_customers', 
                       con = engine, 
                       schema = pg_schema, # pandas is allowing to specify, in which schema the table shall be created
                       if_exists='replace', 
                       index=False
                      )
```
