## Read data: geolocation dataset

```python
df_geolocation=pd.read_csv('../../data/brazilian_e-commerce/olist_geolocation_dataset.csv')
df_geolocation.head()
```

## Data inspection

```python
df_geolocation.info()
```

```python
df_geolocation['geolocation_zip_code_prefix'].nunique()
```

```python
df_geolocation['geolocation_zip_code_prefix'].value_counts()
```

## Data cleaning

```python
df_geolocation['geolocation_zip_code_prefix'].dtype
```

```python
df_geolocation['geolocation_zip_code_prefix'] = (
    df_geolocation['geolocation_zip_code_prefix']
    .astype(str)
    .str.zfill(5)
)
```

```python
df_geolocation['geolocation_zip_code_prefix'].dtype
```

```python
df_geolocation.head()
```

```python
df_geolocation['geolocation_lat'].nunique()
```

```python
df_geo_clean = (
    df_geolocation
    .groupby('geolocation_zip_code_prefix')
    .agg({
        'geolocation_lat': 'mean',
        'geolocation_lng': 'mean'
    })
    .reset_index()
)

df_geo_clean.head()
```

## Merge data: orders_customers + geolocation

```python
df_orders_customers_geo = df_orders_customers.merge(
    df_geo_clean,
    how="left",
    left_on="customer_zip_code_prefix",
    right_on="geolocation_zip_code_prefix"
)
df_orders_customers_geo.head()
```

## Data inspection

```python
df_orders_customers_geo.info()
```

## Data cleaning

```python
df_orders_customers_geo = df_orders_customers_geo.drop(
    columns=['geolocation_zip_code_prefix']
)
```

```python
df_orders_customers_geo.head()
```

```python
import pandas as pd

date_columns = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in date_columns:
    df_orders_customers_geo[col] = pd.to_datetime(
        df_orders_customers_geo[col],
        format='%Y-%m-%d %H:%M:%S',  # typical Olist format
        errors='coerce'               # invalid or missing dates become NaT
    )
```

```python
df_orders_customers_geo.isnull().sum()
```

```python
missing_percent = df_orders_customers_geo.isnull().mean() * 100
missing_percent = missing_percent.sort_values(ascending=False)
print(missing_percent)
```

```python
df_orders_customers_geo = df_orders_customers_geo.dropna(
    subset=['geolocation_lat', 'geolocation_lng']
)
df_orders_customers_geo.head()
```

```python
# Check number of exact duplicate rows
num_duplicates = df_orders_customers_geo.duplicated().sum()
print(f"Number of fully duplicated rows: {num_duplicates}")
```

## Read data: states dataset

```python
df_states=pd.read_csv('../../data/brazilian_e-commerce/states.csv')
df_states.head()
```

## Merge data: orders_customers_geo + states

```python
# Merge customer orders with state info
df_orders_customers_geo_states = pd.merge(
    df_orders_customers_geo,
    df_states,
    how='left',
    left_on='customer_state',  # from orders/customers
    right_on='UF'              # from states table
)
df_orders_customers_geo_states.head()
```

## Data cleaning

```python
# Rename for clarity
df_orders_customers_geo_states.rename(
    columns={
        'geolocation_lat': 'customer_lat',
        'geolocation_lng': 'customer_lng',
        'Latitude': 'state_lat',
        'Longitude': 'state_lng'
    },
    inplace=True
)
df_orders_customers_geo_states.head()
```

```python
# List of unnecessary columns from states table
columns_to_drop = [
    'Area',
    'Demographic Density',
    'Cities count',
    'GDP',
    'GDP rate',
    'Poverty',
    'UF'  # Keep only customer_state for clarity
]

# Drop them
df_orders_customers_geo_states.drop(columns=columns_to_drop, inplace=True)

df_orders_customers_geo_states.head()
```

```python
# Check the result
df_orders_customers_geo_states.info()
```

```python
# Convert all column names to lowercase
df_orders_customers_geo_states.columns = df_orders_customers_geo_states.columns.str.lower()

# Check result
print(df_orders_customers_geo_states.columns)
```

```python
def classify_location(pop):
    if pd.isnull(pop):
        return 'Unknown'  # if population data is missing
    elif pop > 300_000:
        return 'Urban'
    elif 50_000 <= pop <= 300_000:
        return 'City'
    else:  # population < 50_000
        return 'Rural'
```

```python
df_orders_customers_geo_states['location_type'] = df_orders_customers_geo_states['population'].apply(classify_location)
```

```python
df_orders_customers_geo_states['location_type'].value_counts()
```
