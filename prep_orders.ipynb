```python
import pandas as pd

date_columns = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in date_columns:
    df_orders[col] = pd.to_datetime(
        df_orders[col],
        format='%Y-%m-%d %H:%M:%S',  # typical Olist format
        errors='coerce'               # invalid or missing dates become NaT
    )
df_orders.head()
```
